## People Vaccinated / Per Hundred

```python
#Function to label plot points
def label_point(x, y, val, thePlot):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        thePlot.text(point['x']+.02, point['y'], str(point['val']))
        
#Distributed vaccines vs. total at least one dose
distVsTot = sns.lmplot(x='Total Distributed', y='People Vaccinated', 
data=data, height=6, aspect=2, scatter_kws={'color': '#E2FCE6'}, line_kws={'color': '#E2FCE6'})
plt.style.use('dark_background')
plt.title('Total Vaccine Distribution vs Total People Vaccinated')
plt.xlabel('Total Vaccines Distributed')
plt.ylabel('Total Vaccinated at Least Once')
label_point(data['Total Distributed'], data['People Vaccinated'], data['Entity'], plt.gca())  

```

```python
#Distributed vaccines vs. at least one dose per hundred people
distVsTot = sns.lmplot(x='Total Distributed', y='People Vaccinated Per Hundred', 
data=data, height=6, aspect=2, scatter_kws={'color': '#E2FCE6'}, line_kws={'color': '#E2FCE6'})
plt.style.use('dark_background')
plt.title('Total Vaccine Distribution vs Total Per Hundred People Vaccinated')
plt.xlabel('Total Vaccines Distributed')
plt.ylabel('Total per Hundred Vaccinated at Least Once')
label_point(data['Total Distributed'], data['People Vaccinated Per Hundred'], data['Entity'], plt.gca())  
```
